# Emsemble Techniques:
Ensemble leaming is a machine learning paradigm where multiple models (often called "weak learners") are trained to solve the same problem and combined to get better results. The main hypothesis is that when weak models are correctly combined we can obtain more accurate and/or robust models.

âš«bagging, that often considers homogeneous weak learners, learns them independently from each other in parallel and combines them following some kind of deterministic averaging process

âš« boosting, that often considers homogeneous weak learners, learns them sequentially in a very adaptative way (a base model depends on the previous ones) and combines them following a deterministic strategy

âš« stacking, that often considers heterogeneous weak learners, learns them in parallel and combines them by training a meta-model to output a prediction

based on the different weak models predictions

1. Bagging -Random Forest Classifier

2. Boosting -Ada Boost Classifier\Gradient Boosting \Voting classifier \\\XGBoost

3. Stacking Stacking Classifier

Note---->

Bagging and Boosting works with homogeneous (same) models generally DecisionTree

whereas

Stacking uses heterogeneous (different models) at first level and also includes meta model/estimatorÂ atÂ secondÂ level

# Voting Classifier

A Voting Classifier is a machine learning model that trains on an ensemble of numerous models and predicts an output (class) based on their highest probability of chosen class as the output. It simply aggregates the findings of each classifier passed into Voting Classifier and predicts the output class based on the highest majority of voting. The idea is instead of creating separate dedicated models and finding the accuracy for each them, we create a single model which trains by these models and predicts output based on their combined majority of voting for each output class

Voting Classifier supports two types of votings.

Hard Voting: In hard voting, the predicted output class is a class with the highest majority of votes i.e the class which had the highest probability of being predicted by each of the classifiers. Suppose three classifiers predicted the output class(A. A, B), so here the majority predicted A as output. Hence A will be the final prediction.

Soft Voting: In soft voting, the output class is the prediction based on the average of probability given to that class. Suppose given some input to three models, the prediction probability for class A = (0.30, 0.47, 0.53) and B = (0.20, 0.32, 0.40). So the average for class A is 0.4333 and B is 0.3067, the winner is clearly class A because it had the highest probability averaged by eachÂ classifier.

# Pipelines
In the context of machine learning, a pipeline refers to a series of data processing and modeling steps that are chained together in a specific sequence to automate and streamline the machine learning workflow. Pipelines are commonly used to manage and organize the various stages of building, training, and deploying machine learning models. Here's an overview of what a typical machine learning pipeline might include:

Pipelines help automate and streamline these steps by encapsulating each stage into a single unit. # KMeans Clustering
K-means is one of the simplest unsupervised learning algorithms that solves the well known clustering problem. The procedure follows a simple and easy way to classify a given data set through a certain number of clusters (assume k clusters) fixed a priori. The main idea is to define k centres, one for each cluster. These centroids should be placed in a smart way because of different location causes different result. So, the better choice is to place them as much as possible far away from each other. The next step is to take each point belonging to a given data set and associate it to the nearest centroid. When no point is pending, the first step is completed and an early groupage is done. At this point we need to re-calculate k new centroids as barycenters of the clusters resulting from the previous step After we have these k new centroids, a new binding has to be done between the same data set points and the nearest new centroid. A loop hasÂ beenÂ generated



Let's Work with KMeans Clustring
# K-means Overview
Before deving into the dataset, let us briefly discuss how k-means works:

* 1--> The process begins with k centroids intitalised at random.
* 2--> These centroids are used to assign points to its nearest cluster.
* 3--> The mean of all  points within the cluster is then used to update the position of the centroids.
* 4--> The above steps are repeated until the values of the centroids stabilitse.

Hierarchical clustering is a general family of clustering algorithms that build nested clusters by merging or splitting them successively. This hierarchy of clusters is represented as a tree (or dendrogram). The root of the tree is the unique cluster that gathers all the samples, the leaves being the clusters with only one sample.

# Elbo Method :
            SELECTING NUMBER OF CLUSTER K
           
To evalute the preformance of our K-means algorithms we can take a look athe the inertia or objective function values. This is essentially the sum of squared distance our data points are away from their cluster centriod .
By looking at different inerita values for different numbers of clusters(K):

To make minimum distance's data's cluster we used elbo method



This is what inertia evaluates it tells us how far the points within a cluster are. so inertia actually calculates the sum of distance of all the points within a cluster from the centroid of the cluster.

We calculate for all the cluster and the final inertia values is the sum of all these distances. This distance within the clusters is known as intractluster distance. So, inertia gives us the sum of intracluster distance. 

```python
WCSS=[]
K=range(1,15)
for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(df)
    WCSS.append(km.inertia_)
    
plt.plot(K,WCSS,marker='x')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbo Method')
plt.show()
```

```python
# Elbow Method
# K-Means++ : selects initial cluster centers for k-means custring in a smart way to speed up convergence.

from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss,marker='x')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
